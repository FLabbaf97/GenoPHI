```python
import pandas as pd
```

```python
genophi_interaction_matrix = pd.read_csv("/Users/Farzaneh/Documents/phage-bacteria-interaction/GenoPHI/data/interation_matrices/ecoli_interaction_matrix.csv")
our_interaction_matrix = pd.read_csv("/Users/Farzaneh/Documents/phage-bacteria-interaction/phage_interactions/data/multi_source_data_v3/interaction_table.csv")
```

```python
genophi_interaction_matrix.sort_values(by='strain', inplace=True)
genophi_interaction_matrix
```

```python
our_interaction_matrix.loc[our_interaction_matrix['source'] == 'coli_phage_interactions_2023']
coli_phage_interactions_2023 = our_interaction_matrix.loc[our_interaction_matrix['source'] == 'coli_phage_interactions_2023'].copy()
coli_phage_interactions_2023.rename(columns={'bacteria': 'strain'}, inplace=True)
```

```python
coli_phage_interactions_2023['strain'] = coli_phage_interactions_2023['strain'].str.replace('-', '')
coli_phage_interactions_2023['strain'] = coli_phage_interactions_2023['strain'].str.replace('_', '')
coli_phage_interactions_2023['strain'] = coli_phage_interactions_2023['strain'].str.upper()
coli_phage_interactions_2023.sort_values(by='strain', inplace=True)
coli_phage_interactions_2023

```

```python
merged_matrix = pd.merge(
    coli_phage_interactions_2023,
    genophi_interaction_matrix,
    on=['strain', 'phage'],
    how='outer',
    suffixes=('_coli', '_genophi')
)
merged_matrix
```

```python
# Show rows in merged_matrix with at least one null value in any column
contain_null = merged_matrix[merged_matrix.isnull().any(axis=1)]
print("Number of unique strains (bacteria):", merged_matrix['strain'].nunique())
print("Number of unique phages:", merged_matrix['phage'].nunique())
# Show counts for each phage (how many rows for each phage)
print("Phage occurrence counts (merged_matrix):")
print(merged_matrix['phage'].value_counts())
print("\nBacteria occurrence counts (merged_matrix):")
print(merged_matrix['strain'].value_counts())

```

```python
# Compare phages present in coli_phage_interactions_2023 and genophi_interaction_matrix
phages_coli_2023 = set(coli_phage_interactions_2023['phage'].unique())
phages_genophi = set(genophi_interaction_matrix['phage'].unique())

phages_only_in_coli_2023 = phages_coli_2023 - phages_genophi
phages_only_in_genophi = phages_genophi - phages_coli_2023

print("Phages only in coli_phage_interactions_2023 (not in genophi):")
print(sorted(phages_only_in_coli_2023))
print(f"Total: {len(phages_only_in_coli_2023)}")

print("\nPhages only in genophi_interaction_matrix (not in coli_phage_interactions_2023):")
print(sorted(phages_only_in_genophi))
print(f"Total: {len(phages_only_in_genophi)}")
```

```python
# Compare strains present in coli_strain_interactions_2023 and genophi_interaction_matrix
strains_coli_2023 = set(coli_phage_interactions_2023['strain'].unique())
print(len(strains_coli_2023))
strains_genophi = set(genophi_interaction_matrix['strain'].unique())
print(len(strains_genophi))

strains_only_in_coli_2023 = strains_coli_2023 - strains_genophi
strains_only_in_genophi = strains_genophi - strains_coli_2023

strains_common = strains_coli_2023 & strains_genophi
print(len(strains_common))

print("strains only in coli_phage_interactions_2023 (not in genophi):")
print(sorted(strains_only_in_coli_2023))
print(f"Total: {len(strains_only_in_coli_2023)}")

print("\nstrains only in genophi_interaction_matrix (not in coli_phage_interactions_2023):")
print(sorted(strains_only_in_genophi))
print(f"Total: {len(strains_only_in_genophi)}")
```

```python
coli_phage_interactions_2023.reset_index(drop=True, inplace=True)
coli_phage_interactions_2023['interaction'] = coli_phage_interactions_2023['interaction'].apply(lambda x: 0 if (x == 0 or x == '0.0') else 1)
coli_phage_interactions_2023.drop(columns=['source'], inplace=True)
```

```python
outer_merged_matrix = pd.merge(
    coli_phage_interactions_2023,
    genophi_interaction_matrix,
    on=['strain', 'phage'],
    how='outer',
    suffixes=('_coli', '_genophi')
)
print(outer_merged_matrix.shape)
inner_merged_matrix = pd.merge(
    coli_phage_interactions_2023,
    genophi_interaction_matrix,
    on=['strain', 'phage'],
    how='inner',
    suffixes=('_coli', '_genophi')
)
inner_merged_matrix

print(inner_merged_matrix.shape)

# Check if interaction_coli and interaction_genophi are the same in the inner merge
comparison = inner_merged_matrix[['interaction_coli', 'interaction_genophi']].copy()

# Remove rows where either value is missing (to focus on common measurements)
comparison = comparison.dropna(subset=['interaction_coli', 'interaction_genophi'])

# See how many are the same (==, as floats or ints)
same = (comparison['interaction_coli'] == comparison['interaction_genophi'])

num_same = same.sum()
num_total = len(same)

print(f"Number of matching interactions: {num_same} out of {num_total} ({num_same/num_total:.2%})")
print(f"Number of non-matching interactions: {num_total - num_same}")

# Optionally, show non-matching examples:
print("\nExamples of mismatched interactions:")
mismatched = comparison[~same]
print(mismatched.head(10))

```
